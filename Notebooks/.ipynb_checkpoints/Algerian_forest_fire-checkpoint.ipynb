#### Feature selection by using VIF (Variance inflation Factor)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif =pd.DataFrame()
vif["Feature"] =df_copy.columns
```

```python
vif["VIF"] =[variance_inflation_factor(df_copy.values ,i) for i in range(len(df_copy.columns))]
```

```python
df_copy.drop("BUI",axis=1,inplace=True)
vif =pd.DataFrame()
vif["Feature"] =df_copy.columns
vif["VIF"] =[variance_inflation_factor(df_copy.values ,i) for i in range(len(df_copy.columns))]
```

```python
vif
```

```python
df_copy.drop("FFMC",axis=1,inplace=True)
vif =pd.DataFrame()
vif["Feature"] =df_copy.columns
vif["VIF"] =[variance_inflation_factor(df_copy.values ,i) for i in range(len(df_copy.columns))]
vif
```

```python
df_final =df_copy
df_final
```

#### Split the X & y in independant & dependant Variable

```python
X =df_final.drop("FWI",axis=1) ## Independant variables
y =df_final["FWI"] ### Dependant variable
```

```python
X
```

```python
y
```

#### Train model using train test split

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train ,y_test =train_test_split(X,y,test_size=0.20,random_state=42)
print(X_train.shape ,X_test.shape)
```

#### Scale data

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
X_train =scaler.fit_transform(X_train)
X_test =scaler.transform(X_test)

print(X_train)
```

```python
print(X_test)
```

#### Build Linear regression model

```python
from sklearn.linear_model import LinearRegression
reg_model =LinearRegression()
reg_model.fit(X_train ,y_train)
y_pred =reg_model.predict(X_test)
print(y_pred)
```

#### Performance metrics

```python
from sklearn.metrics import mean_squared_error ,mean_absolute_error ,r2_score
mse =mean_squared_error(y_test ,y_pred)
mae =mean_absolute_error(y_test ,y_pred)
r2_score =r2_score(y_test ,y_pred)
print(f"mean square error=> {mse}\nmean absolute error=> {mae}\nr2-score=> {r2_score}")
```

#### Ridge Regression 

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error ,r2_score
ridge_model =Ridge()
ridge_model.fit(X_train ,y_train)
ridge_ypred =ridge_model.predict(X_test)

print("Ridge Mean square error =>",mean_squared_error(y_test ,ridge_ypred))
print("Ridge R2-score =>",r2_score(y_test ,ridge_ypred))
```

#### Ridge Cross Validation 

```python
from sklearn.linear_model import RidgeCV
ridge_cv =RidgeCV(cv=5)
ridge_cv.fit(X_train ,y_train)
ridgeCV_ypred =ridge_cv.predict(X_test)

print("RidgeCV Mean square error =>",mean_squared_error(y_test ,ridgeCV_ypred))
print("RidgeCV R2-score =>",r2_score(y_test ,ridgeCV_ypred))
```

#### Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso_model =Lasso()
lasso_model.fit(X_train ,y_train)
lasso_ypred =lasso_model.predict(X_test)

print("Lasso Mean square error =>",mean_squared_error(y_test ,lasso_ypred))
print("Lasso R2-score =>",r2_score(y_test ,lasso_ypred))
```

#### Lasso Cross Validation

```python
from sklearn.linear_model import LassoCV
lassoCV_model =LassoCV(cv=5)
lassoCV_model.fit(X_train ,y_train)
lasso_y_pred =lassoCV_model.predict(X_test)


print("LassoCV Mean square error =>",mean_squared_error(y_test ,lasso_y_pred))
print("LassoCV R2-score =>",r2_score(y_test ,lasso_y_pred))
```

#### Elastic Net regression

```python
from sklearn.linear_model import ElasticNet
elasticnet_model =ElasticNet()
elasticnet_model.fit(X_train ,y_train)
elasticnet_y_pred =elasticnet_model.predict(X_test)

print("ElasticNet Mean square error =>",mean_squared_error(y_test ,elasticnet_y_pred))
print("ElasticNet R2-score =>",r2_score(y_test ,elasticnet_y_pred))
```

#### ElasticNet Cross validation

```python
from sklearn.linear_model import ElasticNetCV
elasticnetCV_model =ElasticNetCV(cv=5)
elasticnetCV_model.fit(X_train ,y_train)
elasticnetCV_y_pred =elasticnetCV_model.predict(X_test)

print("ElasticNet Mean square error =>",mean_squared_error(y_test ,elasticnetCV_y_pred))
print("ElasticNet R2-score =>",r2_score(y_test ,elasticnetCV_y_pred))
```
